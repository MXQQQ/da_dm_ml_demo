```python
from urllib.request import urlopen
url = 'http://baidu.com'
resp = urlopen(url)
print(type(resp))
print(resp.status, resp.reason)
print(resp.geturl())
print(resp.info())
print(resp.read())
```

```python
import urllib.parse
import urllib3
dic = {}
dic['wd'] = '张三'
endic = urllib.parse.urlencode(dic)
url = 'https://www.baidu.com/s?{0}'.format(endic)
uesr_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.67 Safari/537.36 Edg/87.0.664.52'
with urllib3.PoolManager() as http:
    resp = http.request('GET', url,headers={'User-Agent':uesr_agent})
    print(type(resp))
    print(resp.status, resp.reason)
    print(resp.geturl())
    print(resp.info())
    print(resp.data)
```
